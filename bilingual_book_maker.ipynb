```python
#@title 上傳epub檔案
from google.colab import files
uploaded = files.upload()
```

```python
#@title Billingual book maker 執行epub翻譯 { vertical-output: true }

#@markdown <b>設定底下的自訂參數後，就可以按左側的執行鈕</b>

#@markdown 你的openai api key，請至https://platform.openai.com/account/api-keys 建立後貼於此
openai_key="你的openai api key" #@param {type:"string"}
#@markdown 翻譯的目標語言代碼
language = 'Traditional Chinese' #@param ["Traditional Chinese", "English", "Simplified Chinese", "Japanese", "Korean", "German", "French", "Spanish" , "Thai", "Ukrainian", "Indonesian", "Finnish", "Dutch", "Turkish", "Greek", "Italian", "Swedish", "Arabic"]
#@markdown 使用的模型
model = 'chatgpt' #@param ['chatgpt', 'gpt3']

#@markdown 是否開啟測試，只翻譯前10筆內容，可先試翻再取消勾選後翻全本
test = True #@param {type: 'boolean'}

#@markdown 設定欲測試翻譯的內容數量，預設為10筆
test_num = 10 #@param {type: 'integer'}

#@markdown 是否全部翻譯完成，立即下載翻譯結果
start_downloading_immediately = True #@param { type: 'boolean' }

#@markdown 是否開啟無限制處理，若有開啟付費或是想把免費的額度用完可以開啟
no_limit = True  #@param { type: 'boolean' }

#@markdown ---
import os 
import subprocess
import google

if uploaded:
  filename = list(uploaded.keys())[0]
  filesize = len(uploaded[filename])
  print(f"Uploaded file: {filename}, File size(bytes): {filesize}")
else:
  filename = "test_books/animal_farm.epub"
  print(f"Use default file: {filename}")

cmd = [
    'python3',
    'make_book.py',
    '--book_name', filename,
    '--openai_key', openai_key,
    '--language', language,
    '--model', model
]

if test:
  cmd.append("--test")
  if test_num !=10:
    cmd.append("--test_num")
    cmd.append(test_num)
if no_limit:
  cmd.append("--no_limit")

result = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

print(result.stdout.decode())
print(result.stderr.decode())
filename_stem= filename.split(".")[0]
exported_filename = f"{filename_stem}_bilingual.epub"


# Get translated result
if os.path.exists(exported_filename) :
  if start_downloading_immediately:
    google.colab.files.download(exported_filename)

else :
    print('找不到翻譯結果: '+exported_filename)
    start_downloading_immediately = False

```
